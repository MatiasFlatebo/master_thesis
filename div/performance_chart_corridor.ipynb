```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
horizons = [16, 24, 32, 40, 48]
dp_scores = [0.54, 0.60, 0.50, 0.64, 0.60]
sdp_scores = [0.60, 0.69, 0.84, 0.96, 1.0]
```

```python
sns.set_theme(style="whitegrid", context="talk", font_scale=1.2)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(horizons, dp_scores, marker='o', label='DP', linewidth=2)
plt.plot(horizons, sdp_scores, marker='o', label='SDP', linewidth=2)

# Labels and title
plt.xlabel("Prediction Horizon", labelpad=10)
plt.ylabel("Score", labelpad=10)
plt.title("Corridor Environment: Performance by Prediction Horizon", pad=15, fontsize=16, weight='bold')

# Legend and axis
plt.legend(frameon=False)
plt.xticks(horizons)
plt.ylim(0, 1.1)

# Remove top/right spines for clean look
sns.despine()

# Final touches
plt.tight_layout()
plt.savefig(f"images/chart_corridor.png", dpi=300)
plt.show()
```
